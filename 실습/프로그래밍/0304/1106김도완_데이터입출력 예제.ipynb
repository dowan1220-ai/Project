```python
pen = int(input("연필의 판매 갯수는? "))
pen_price = 700
eraser = int(input("지우개의 판매 갯수는? "))
eraser_price = 500
notbook = int(input("공책의 판매 갯수는? "))
notbook_price = 2500

total = pen_price * pen + eraser * eraser_price + notbook * notbook_price

print("---------------------------")
print("파이썬 문방구의 오늘 판매 내역")
print("---------------------------")
print(f"1. 연필  : {pen}개, {pen * pen_price} 원")
print(f"2. 지우개 : {eraser}개, {eraser * eraser_price} 원")
print(f"3. 공책  : {notbook}개, {notbook * notbook_price} 원")
print("---------------------------")
print(f"오늘 총 판매 금액 : {total} 원")
```
